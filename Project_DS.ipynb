```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
movies = pd.read_csv("/content/movies.csv")
ratings = pd.read_csv("/content/ratings.csv", nrows=500000)
movies.head()
ratings.head()
data = pd.merge(ratings, movies, on="movieId")
data.head()
data.groupby("title")["rating"].mean().sort_values(ascending=False).head(10)
data["title"].value_counts().head(10)
plt.hist(data["rating"], bins=10)
plt.title("Distribution of Ratings")
plt.xlabel("Ratings")
plt.ylabel("Frequency")
plt.show()
data["title"].value_counts().head(10).plot(kind="bar")

plt.title("Top 10 Most Rated Movies")
plt.xlabel("Movies")
plt.ylabel("Number of Ratings")

plt.show()
```
